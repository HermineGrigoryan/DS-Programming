## Exercise 1

ls = [1, 2, 3, 0]

Which of the following expressions will output 'None'?

1. sorted(ls) 
2. ls.append('5')
3. ls.extend(None)
4. ls + "some_string" 
5. ls = ls.append(1)

```
ls = [1, 2, 3, 0]
print(ls, id(ls))
print(sorted(ls))
ls.append(['5', '7'])
print(ls, id(ls))
ls = ls + ["some_string"]
print(ls, id(ls))
ls.extend(list(('another_string', 9)))
print(ls, id(ls))
ls = ls.append([1, 2]) # this expression will output None as 'append' is an in-place method 
print(ls, id(ls))
```

As we can see, the id-s of the first 2 lists are the same because we didn't create a new object, whereas in the third expression, we created a new object.
Pay attention that by using 'append' we are appending a list to another list and getting a list of lists. In case of extend, we are adding the values of the second list to the first string.

## Exercise 2

You have the following list: old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 'a']].

Change the second number from the third sublist to 'changed_number'. Before doing this operation, assign old_list to new_list using 1. alising, 2. copy(), 3. deepcopy().
Will the results be different for these three cases?

```
# Aliasing - if we change new_list, old_list will be changed as well
old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 'a']]
new_list = old_list

new_list[2][1] = 'changed_number'

print('Old List:', old_list)
print('ID of Old List:', id(old_list))

print('New List:', new_list)
print('ID of New List:', id(new_list))
```

```
# Shallow aliasing (copy) - if we change the state of underlying values inside the sublists in new_list, the old_list will be changed as well. 
old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 'a']]
new_list = old_list.copy()

new_list[2][1] = 'changed_number'

print('Old List:', old_list)
print('ID of Old List:', id(old_list))

print('New List:', new_list)
print('ID of New List:', id(new_list))
```

```
# Shallow aliasing (copy) - if we are not changing the state of underlying values, old_list will not be altered. 
old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 'a']]
new_list = old_list.copy()

new_list.append('changed_number')

print('Old List:', old_list)
print('ID of Old List:', id(old_list))

print('New List:', new_list)
print('ID of New List:', id(new_list))
```

```
# In case of deep copy, we copy not only the references of a list but also we copy the underlying values.
# Therefore, while mutating new_list, old_list is not changed.
import copy
old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 'a']]
new_list = copy.deepcopy(old_list)

new_list[2][1] = 'changed_number'

print('Old List:', old_list)
print('ID of Old List:', id(old_list))

print('New List:', new_list)
print('ID of New List:', id(new_list))
```

## Exercise 2

Calculate how much money you spent in the supermarket using the below dictionary. The first element in each list represents the number of items, and the second element represents the price for one item.

```
cart  = {
    "banana": [8, 2],
    "apple": [2, 5],
    "orange": [3, 1.5],
    "pear": [3, 6]
  }
```

```
cart['banana'][0]*cart['banana'][1]+cart['apple'][0]*cart['apple'][1]+cart['orange'][0]*cart['orange'][1]+cart['pear'][0]*cart['pear'][1]
```

```
total = 0
for i in cart.keys():
  total += cart[i][0]*cart[i][1]
total
```

```
total = 0

for i in range(len(cart)):
  total += list(cart.values())[i][0]*list(cart.values())[i][1]
total
```

```
total = 0
for i, element in enumerate(cart.values()):
  total += element[0]*element[1]
total
```

```
sum(
    map(
      lambda val: val[0] * val[1],
      cart.values()
  )
)
```

## Exercise 1

You have three lists. 
* Find the common numbers in these lists.
* Check which of them are binary lists (contain only 0 and 1).

```
l1 = [1, 0, 1, 5, 6, 80]
l2 = [1, 0, 0, 1, 0]
l3 = ['one', 5, 0, 1, 15]

set(l1).intersection(l2).intersection(l3) # set(l1) & set(l2) & set(l3)
```

```
binary_set = {0, 1}
print(set(l1).union(binary_set)==binary_set) # |
print(set(l2).union(binary_set)==binary_set)
print(set(l3).union(binary_set)==binary_set)
```

```
s1 = "the big dwarf only jumps"
s2 = "noon"
s3 = "mama"
s4 = "most probably this string is neither a heterogram, nor an isogram"
```

```
len(set(s1)) == len(s1) #remove the punctuation
s1 = s1.replace(' ', '')
print('Heterogram:', len(set(s1)) == len(s1)) 
print('Heterogram:', len(set(s2)) == len(s2)) 
print('Heterogram:', len(set(s3)) == len(s3))
print('Heterogram:', len(set(s4)) == len(s4))  
```
