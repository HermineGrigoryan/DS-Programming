# OOP Overview

**Object-oriented programming (OOP)** is a method of structuring a program by bundling related properties and behaviors into individual objects. For instance, an object could represent a person with properties like a name, age, and address and behaviors such as walking, talking, breathing, and running. Or it could represent an email with properties like a recipient list, subject, and body and behaviors like adding attachments and sending.

Create a complete class of `University`.

```python
class University:
    
    '''This class is about Universities.'''
    
    # class variable
    institution_type = 'University'
    n_uni = 0 
    tuition_perc_change = 1.2
    
    # instance variables unique to each instance (__init__ constructor)
    def __init__(self, uni_name, country, uni_ranking, tuition): #instance variables
        self.uni_name = uni_name
        self.uni_ranking = uni_ranking
        self.country = country
        self.tuition = tuition
        
        University.n_uni += 1
        
    def uni_country(self):
        return f'{self.uni_name} is in {self.country}'
    
    def name_ranking(self):
        return f'{self.uni_name} has {self.uni_ranking} ranking'
    
    def official_website(self):
        return f'www.{self.uni_name.lower()}.org'
    
    def change_tuition(self):
        self.tuition = int(self.tuition*self.tuition_perc_change)
```

## Lecturer subclass
Add new attributes and methods, such as:

```python
# Create a Lecturer subclass
class Lecturer(University):
    
    '''This is a Lecturer subclass of University class'''
    
    # adding new attributes to the attributes in the University class
    def __init__(self, uni_name, country, uni_ranking, tuition, lect_name, subjects = None):
        super().__init__(uni_name, country, uni_ranking, tuition)
        self.lect_name = lect_name
        self.lect_first_name, self.lect_last_name = self.lect_name.split(' ')
        self.subjects = subjects
    
    def lecturer_email(self):    
        return f'{self.stud_first_name}.{self.stud_last_name}@{self.uni_name}.com'
    
    # adding new subjects to the list of subjects
    def add_subjects(self, subject):
        if subject not in self.subjects:
            self.subjects.extend(subject)
    
```

```python
# Link students to the professors. 
# E.g., stud1 and stud3 are lect2 students, stud2 is lect3's student.
# For that, create add_students and print_students methods.
class Lecturer(University):
    
    '''This is a Lecturer subclass of University class'''
    
    # adding new attributes to the attributes in the University class
    def __init__(self, uni_name, country, uni_ranking, tuition, lect_name, subjects = None, students = None):
        super().__init__(uni_name, country, uni_ranking, tuition)
        self.lect_name = lect_name
        self.lect_first_name, self.lect_last_name = self.lect_name.split(' ')
        self.subjects = subjects
        self.students = students
    
    def lecturer_email(self):    
        return f'{self.stud_first_name}.{self.stud_last_name}@{self.uni_name}.com'
    
    # adding new subjects to the list of subjects
    def add_subjects(self, subject):
        if subject not in self.subjects:
            self.subjects.extend(subject)
    
    # adding new students to the list of students
    def add_students(self, student):
        if student not in self.students:
            self.students.extend(student)
    
    def print_students(self):
        if self.students is not None:
            for emp in self.students:
                print(self.lect_name, ':', emp.stud_name)
```
